## Описание данных

Данные состоят из файлов, полученных из разных источников:

- `data_arc.csv` — данные об электродах;
- `data_bulk.csv` — данные о подаче сыпучих материалов (объём);
- `data_bulk_time.csv` *—* данные о подаче сыпучих материалов (время);
- `data_gas.csv` — данные о продувке сплава газом;
- `data_temp.csv` — результаты измерения температуры;
- `data_wire.csv` — данные о проволочных материалах (объём);
- `data_wire_time.csv` — данные о проволочных материалах (время).

Во всех файлах столбец key содержит номер партии. В файлах может быть несколько строк с одинаковым значением key: они соответствуют разным итерациям обработки.

Цель проекта - уменьшить потребление электроэнергии на этапе обработки стали путем построения модели, предсказывающей температуру.

### Описание этапа обработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

## Предобработка данных

Даты в наших таблицах загружены корректно, остальные столбцы тоже имеют корректный тип, за исключением температуры из data_temp. Исправим, а заодно удалим строки с пропущенными значениями, которые не нужны нам при построении модели.

```python
data_temp.dropna(inplace=True) # удаляем пропуски
# data_temp = data_temp[data_temp['Температура'] > 1300] # удаляем выбросы с крайне низкой температурой

count = data_temp.groupby(by='key', as_index=False).count() # создадим переменную с группировкой по номеру партии
count = count[count['Время замера']== 1]['key'].tolist() # создадим список с партиями, у которых был всего один замер
data_temp = data_temp.query('key not in @count') # исключим их из основной таблицы
del count

data_temp['Температура'] = data_temp['Температура'].astype(int) # приводим к нужному типу
data_temp.info() # проверяем
```

```python
# приводим наименование столбцов к одному языку и змеиному_регистру
data_arc.columns = ['key', 'start_time','end_time', 'active_power', 'reactive_power', 'full_power']
data_bulk.columns = ['key', 'bulk_1','bulk_2','bulk_3','bulk_4','bulk_5','bulk_6','bulk_7','bulk_8',
                     'bulk_9','bulk_10','bulk_11','bulk_12','bulk_13','bulk_14','bulk_15']
data_bulk_time.columns = ['key', 'bulk_1','bulk_2','bulk_3','bulk_4','bulk_5','bulk_6','bulk_7','bulk_8',
                          'bulk_9','bulk_10','bulk_11','bulk_12','bulk_13','bulk_14','bulk_15']
data_gas.columns = ['key', 'gas_1']
data_temp.columns = ['key', 'time','temp']
data_wire.columns = ['key','wire_1','wire_2','wire_3','wire_4','wire_5','wire_6','wire_7','wire_8','wire_9']
data_wire_time.columns = ['key','wire_1','wire_2','wire_3','wire_4','wire_5','wire_6','wire_7','wire_8','wire_9']
```

```python
# заполним пропуски в сыпучих и проволочных материалах на 0
data_bulk = data_bulk.fillna(0)
data_wire = data_wire.fillna(0)
data_bulk_time = data_bulk_time.fillna(0)
data_wire_time = data_wire_time.fillna(0)

data_wire.drop('wire_5', axis=1, inplace=True) # заодно удалим столбец wire 5, т.к. он пустой
data_wire_time.drop('wire_5', axis=1, inplace=True)
```

```python
data_temp = data_temp.groupby(by='key', as_index = False).agg(['first', 'last']) # группируем по номеру партии
data_temp = data_temp.droplevel(0, axis=1).rename_axis(None, axis=1).reset_index() # убираем двухэтажное наименование столбцов
data_temp.columns = ['key', 'start_time', 'end_time', 'temp_first', 'temp_last'] # переименовываем их
data_temp.head() # проверяем
```

```python
data_arc['heating_time'] = (data_arc['end_time'] - data_arc['start_time']).dt.seconds # вычисляем время нагрева в секундах
data_arc = data_arc[['key', 'full_power', 'heating_time']] # оставляем только нужные столбцы
data_arc = data_arc.groupby('key').agg({'full_power': 'sum', 'heating_time': 'mean'})
data_temp.drop('start_time', axis=1, inplace=True)
data_temp.drop('end_time', axis=1, inplace=True)
```

```python
# Создаем итоговую таблицу
data = data_temp.merge(right=data_arc,
                       how='inner',
                       on='key')\
                .merge(right=data_bulk,
                       how='inner',
                       on='key')\
                .merge(right=data_gas,
                       how='inner',
                       on='key')\
                .merge(right=data_wire,
                       how='inner',
                       on='key')
display(data.head())
data.info()
```

Итоговый датафрейм был объединен по столбцу `key`, поэтому не имеет пропусков. Столбцы в нем имеют корректные наименования и нужный тип.

## Обучение моделей

В данном шаге нам необходимо обучить три модели:
* Линейную регрессию;
* Случайный лес;
* CatBoost.

Для этого создадим переменные с признаками и целевым признаком и разобьем данные на тренировочный и тестовый наборы в соотношении 75:25.
Также, я считаю, необходимо применить масштабирование, дабы одни признаки не доминировали над другими по величине. Иначе модель не будет учитывать вклад переменных меньшего масштаба, даже если он сильный.

```python
data.drop('key', axis=1, inplace=True) # Исключаем key из датасета, т.к. это просто номер партии
```

```python
# создаем переменные с признаками и целевым признаком
target = data['temp_last']
features = data.drop(['temp_last'], axis=1)

# производим деление на выборки в соотношении 3 к 1
(features_train,
 features_test,
 target_train,
 target_test) = train_test_split(features, target, test_size=0.25, random_state=231222)

# проверяем
print(features_train.shape)
print(features_test.shape)
print(target_train.shape)
print(target_test.shape)
```

```python
data
```

```python
# добавляю масштабирование признаков
scaler = StandardScaler()
scaler.fit(features_train)
features_train = scaler.transform(features_train)
features_test = scaler.transform(features_test)
```

### Линейная регрессия

```python
%%time

model_LR = LinearRegression()

MAE_LR = (cross_val_score(model_LR, 
                          features_train, 
                          target_train, 
                          cv=5, 
                          scoring='neg_mean_absolute_error').mean()*-1)
print('LR MAE:', MAE_LR)
```

### Случайный лес

```python
%%time

model_RF = RandomForestRegressor() 
hyperparams = [{'criterion':['mse'],
                'max_depth': [10, 20, 30, 40, 50],
                'n_estimators': [50, 100, 200, 300], 
                'random_state': [231222]}]
clf = GridSearchCV(model_RF, hyperparams, scoring='neg_mean_absolute_error', cv=5)
clf.fit(features_train, target_train)

best_params = clf.best_params_
print('Параметры:', clf.best_params_)
means = clf.cv_results_['mean_test_score']

MAE_RFR = (min(means)*-1)
print('RF MAE:', MAE_RFR)
```

### CatBoost

```python
%%time

model_cat = CatBoostRegressor(verbose=False, random_state=231222)

MAE_cat = (cross_val_score(model_cat, 
                           features_train, 
                           target_train, 
                           cv=5, 
                           scoring='neg_mean_absolute_error').mean() * -1)
print('cat MAE', MAE_cat)
best_params_cat = CatBoostRegressor(verbose=False, 
                                    random_state=231222).fit(features_train, target_train).get_all_params()
print()
print('Параметры:', best_params_cat)
```

### Вывод по моделям

```python
index = ['LinearRegression',
         'RandomForestRegressor',
         'CatBoostRegressor']

data_t = {'MAE':[MAE_LR, MAE_RFR, MAE_cat]}
pivot_data = pd.DataFrame(data=data_t, index=index)
pivot_data
```

Лучший результат в `6.02` показывает CatBoost, поэтому тест будем проводить именно на нем.

## Тестирование

```python
%%time

model_cat = CatBoostRegressor(verbose=False, random_state=231222)
model_cat.set_params(**best_params_cat)
model_cat.fit(features_train, target_train)
cat_predict = model_cat.predict(features_test)
test_MAE_cat = mean_absolute_error(cat_predict, target_test)
print('MAE cat', test_MAE_cat)
```

```python
# Построим график важности признаков

fig, ax = plt.subplots(figsize=(15,10))
sort = model_cat.feature_importances_.argsort()
plt.barh(features.columns.values[sort], model_cat.feature_importances_[sort])
_ = ax.set_title('Оценка важности признаков')
_ = ax.set_ylabel('Признаки')
```

Видим по графику важности факторов, что больше всего на наш целевой признак влияет:
* Температура в начале;
* gas_1;
* bulk_11, _15, _13, _14

А признаки wire_2, _6, _7 и bulk_1, _7, _12 не оказывают на нее никакого влияния.
